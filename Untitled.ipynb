```python
import pandas as pd
import matplotlib.pyplot as plt

age_survival = train_df.groupby("Age")["Survived"].mean()

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(age_survival.index, age_survival.values, marker='o', linestyle='-')
plt.title("Survival Rate by Age")
plt.xlabel("Age")
plt.ylabel("Survival Rate")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot to show survival distribution by sex
sns.countplot(x="Sex", hue="Survived", data=train_df)
plt.title("Survival Counts by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show
```
